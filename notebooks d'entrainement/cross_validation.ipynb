```python
# ================== Imports ==================
import os, torch, torch.nn as nn, torch.nn.functional as F
from torchvision import models, transforms, datasets
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    confusion_matrix,
    ConfusionMatrixDisplay
)

torch.backends.cudnn.benchmark = True

# ================== Configuration ==================
class Config:
    img_size = 224
    patch_size = 16
    embed_dim = 768
    num_heads = 12
    num_layers = 6
    dropout = 0.1
    efficientnet_features = 1280  # sortie des features d'EfficientNet-B0
    lr = 1e-4
    weight_decay = 1e-2
    batch_size = 32
    epochs =9

# ====== Chemins locaux ======
PROJECT_ROOT = os.getcwd()
TRAIN_DIR = os.path.join(PROJECT_ROOT, "train")
VAL_DIR   = os.path.join(PROJECT_ROOT, "val")
TEST_DIR  = os.path.join(PROJECT_ROOT, "test")

print("üìÇ R√©pertoire courant :", PROJECT_ROOT)
print("üì∏ Dossiers trouv√©s :", os.listdir(PROJECT_ROOT))

# ================== Vision Transformer Hybride ==================
class PatchEmbedding(nn.Module):
    def __init__(self, feature_dim, embed_dim, num_patches):
        super().__init__()
        self.projection = nn.Linear(feature_dim, embed_dim)
        self.position_embeddings = nn.Parameter(torch.randn(1, num_patches + 1, embed_dim))
        self.cls_token = nn.Parameter(torch.randn(1, 1, embed_dim))

    def forward(self, x):
        b = x.shape[0]
        cls_tokens = self.cls_token.expand(b, -1, -1)
        x = self.projection(x)
        x = torch.cat([cls_tokens, x], dim=1)
        x = x + self.position_embeddings
        return x

class MultiHeadAttention(nn.Module):
    def __init__(self, embed_dim, num_heads, dropout=0.1):
        super().__init__()
        self.embed_dim = embed_dim
        self.num_heads = num_heads
        self.head_dim = embed_dim // num_heads
        assert self.head_dim * num_heads == embed_dim
        self.qkv = nn.Linear(embed_dim, embed_dim * 3)
        self.out_proj = nn.Linear(embed_dim, embed_dim)
        self.dropout = nn.Dropout(dropout)

    def forward(self, x):
        b, L, d = x.shape
        qkv = (
            self.qkv(x)
            .reshape(b, L, 3, self.num_heads, self.head_dim)
            .permute(2, 0, 3, 1, 4)
        )
        q, k, v = qkv[0], qkv[1], qkv[2]
        attn = (q @ k.transpose(-2, -1)) / (self.head_dim ** 0.5)
        attn = torch.softmax(attn, dim=-1)
        attn = self.dropout(attn)
        out = (attn @ v).transpose(1, 2).reshape(b, L, d)
        return self.out_proj(out)

class TransformerBlock(nn.Module):
    def __init__(self, embed_dim, num_heads, mlp_ratio=4.0, dropout=0.1):
        super().__init__()
        self.norm1 = nn.LayerNorm(embed_dim)
        self.attn = MultiHeadAttention(embed_dim, num_heads, dropout)
        self.norm2 = nn.LayerNorm(embed_dim)
        hidden = int(embed_dim * mlp_ratio)
        self.mlp = nn.Sequential(
            nn.Linear(embed_dim, hidden),
            nn.GELU(),
            nn.Dropout(dropout),
            nn.Linear(hidden, embed_dim),
            nn.Dropout(dropout),
        )

    def forward(self, x):
        x = x + self.attn(self.norm1(x))
        x = x + self.mlp(self.norm2(x))
        return x

class HybridEfficientNetViT(nn.Module):
    def __init__(self, num_classes, cfg: Config):
        super().__init__()
        from torchvision.models import EfficientNet_B0_Weights
        eff = models.efficientnet_b0(weights=EfficientNet_B0_Weights.DEFAULT)
        self.backbone = eff.features
        self.pool = nn.AdaptiveAvgPool2d((7, 7))
        num_patches = 49
        self.patch_embed = PatchEmbedding(cfg.efficientnet_features, cfg.embed_dim, num_patches)
        self.blocks = nn.ModuleList([
            TransformerBlock(cfg.embed_dim, cfg.num_heads, dropout=cfg.dropout)
            for _ in range(cfg.num_layers)
        ])
        self.norm = nn.LayerNorm(cfg.embed_dim)
        self.cls_head = nn.Sequential(
            nn.Dropout(cfg.dropout),
            nn.Linear(cfg.embed_dim, num_classes)
        )

    def forward(self, x):
        feats = self.backbone(x)
        feats = self.pool(feats)
        feats = feats.flatten(2).transpose(1, 2)
        x = self.patch_embed(feats)
        for blk in self.blocks:
            x = blk(x)
        x = self.norm(x)
        cls_tok = x[:, 0]
        return self.cls_head(cls_tok)

# ================== Chargement des donn√©es ==================
def get_transforms():
    train_tf = transforms.Compose([
        transforms.Resize((Config.img_size, Config.img_size)),
        transforms.RandomResizedCrop(Config.img_size, scale=(0.85, 1.0)),
        transforms.RandomHorizontalFlip(),
        transforms.RandomVerticalFlip(),
        transforms.RandomRotation(25),
        transforms.ColorJitter(0.2, 0.2, 0.2, 0.1),
        transforms.ToTensor(),
        transforms.Normalize([0.485,0.456,0.406],[0.229,0.224,0.225]),
    ])
    eval_tf = transforms.Compose([
        transforms.Resize((Config.img_size, Config.img_size)),
        transforms.ToTensor(),
        transforms.Normalize([0.485,0.456,0.406],[0.229,0.224,0.225]),
    ])
    return train_tf, eval_tf

def make_dataloaders():
    train_tf, eval_tf = get_transforms()
    train_ds = datasets.ImageFolder(TRAIN_DIR, transform=train_tf)
    val_ds   = datasets.ImageFolder(VAL_DIR,   transform=eval_tf)
    test_ds  = datasets.ImageFolder(TEST_DIR,  transform=eval_tf)

    num_classes = len(train_ds.classes)
    print(f"Classes d√©tect√©es ({num_classes}) : {train_ds.classes[:10]} ...")

    train_loader = DataLoader(train_ds, batch_size=Config.batch_size, shuffle=True, num_workers=0)
    val_loader   = DataLoader(val_ds, batch_size=Config.batch_size, shuffle=False, num_workers=0)
    test_loader  = DataLoader(test_ds, batch_size=Config.batch_size, shuffle=False, num_workers=0)

    return train_loader, val_loader, test_loader, train_ds.classes, num_classes

# ================== Entra√Ænement ==================
def train_model(model, train_loader, val_loader, epochs=Config.epochs, device="cuda"):
    model = model.to(device)
    criterion = nn.CrossEntropyLoss()
    optimizer = torch.optim.AdamW(model.parameters(), lr=Config.lr, weight_decay=Config.weight_decay)
    scheduler = torch.optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=epochs)
    best = 0.0
    scaler = torch.cuda.amp.GradScaler(enabled=(device=="cuda"))

    for ep in range(1, epochs+1):
        model.train()
        run_loss, correct, n = 0.0, 0, 0
        for x, y in train_loader:
            x, y = x.to(device), y.to(device)
            optimizer.zero_grad(set_to_none=True)
            with torch.cuda.amp.autocast(enabled=(device=="cuda")):
                logits = model(x)
                loss = criterion(logits, y)
            scaler.scale(loss).backward()
            scaler.step(optimizer)
            scaler.update()
            run_loss += loss.item()
            pred = logits.argmax(1)
            correct += (pred==y).sum().item()
            n += y.size(0)

        # Validation
        model.eval()
        v_loss, v_corr, m, y_true, y_pred = 0.0, 0, 0, [], []
        with torch.no_grad():
            for x, y in val_loader:
                x, y = x.to(device), y.to(device)
                logits = model(x)
                loss = criterion(logits, y)
                v_loss += loss.item()
                p = logits.argmax(1)
                y_true.extend(y.cpu().numpy())
                y_pred.extend(p.cpu().numpy())
                v_corr += (p==y).sum().item()
                m += y.size(0)

        train_acc = 100*correct/max(1,n)
        val_acc   = 100*v_corr/max(1,m)
        precision = 100*precision_score(y_true, y_pred, average="macro", zero_division=0)
        recall    = 100*recall_score(y_true, y_pred, average="macro", zero_division=0)
        f1        = 100*f1_score(y_true, y_pred, average="macro", zero_division=0)

        print(f"\nEpoch {ep}/{epochs}")
        print(f"Train - Loss:{run_loss/len(train_loader):.4f}  Acc:{train_acc:.2f}%")
        print(f"Val   - Loss:{v_loss/len(val_loader):.4f}  Acc:{val_acc:.2f}%  P:{precision:.2f}%  R:{recall:.2f}%  F1:{f1:.2f}%")

        if val_acc > best:
            best = val_acc
            torch.save(model.state_dict(), os.path.join(PROJECT_ROOT, "best_hybrid_plantvillage.pth"))
            print(f"‚úÖ Nouveau meilleur mod√®le sauvegard√© (Val Acc {val_acc:.2f}%).")
        scheduler.step()
    return model

# ================== √âvaluation ==================
def evaluate(model, loader, class_names, device="cuda", title="Matrice de confusion"):
    model.eval()
    y_true, y_pred = [], []
    with torch.no_grad():
        for x, y in loader:
            x = x.to(device)
            logits = model(x)
            p = logits.argmax(1).cpu().numpy()
            y_pred.extend(p)
            y_true.extend(y.cpu().numpy())
    cm = confusion_matrix(y_true, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=class_names)
    disp.plot(cmap="Blues", xticks_rotation=45)
    plt.title(title)
    plt.tight_layout()
    plt.show()
    acc = 100*accuracy_score(y_true, y_pred)
    macro_f1 = 100*f1_score(y_true, y_pred, average="macro")
    print(f"Test Acc: {acc:.2f}% | Macro-F1: {macro_f1:.2f}%")

# ================== Lancement ==================
device = "cuda" if torch.cuda.is_available() else "cpu"
print("‚öôÔ∏è  Appareil utilis√© :", device)

train_loader, val_loader, test_loader, class_names, num_classes = make_dataloaders()

model = HybridEfficientNetViT(num_classes=num_classes, cfg=Config())
print(f"Param√®tres du mod√®le : {sum(p.numel() for p in model.parameters()):,}")

model = train_model(model, train_loader, val_loader, device=device)

# Chargement du meilleur mod√®le
model.load_state_dict(torch.load(os.path.join(PROJECT_ROOT, "best_hybrid_plantvillage.pth"), map_location=device))
model = model.to(device)
evaluate(model, test_loader, class_names=class_names, device=device, title="Matrice de confusion (Test)")
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, classification_report
from collections import Counter

def visualize_model_performance(model, loader, class_names, device="cuda"):
    model.eval()
    y_true, y_pred = [], []

    with torch.no_grad():
        for x, y in loader:
            x = x.to(device)
            preds = model(x).argmax(1).cpu().numpy()
            y_pred.extend(preds)
            y_true.extend(y.numpy())

    # ===== Matrice de confusion =====
    cm = confusion_matrix(y_true, y_pred)
    cm_norm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]

    plt.figure(figsize=(14, 10))
    sns.heatmap(cm_norm, annot=False, cmap="Blues", cbar=True)
    plt.title("üåø Matrice de confusion normalis√©e (Test)", fontsize=16)
    plt.xlabel("Classe pr√©dite", fontsize=13)
    plt.ylabel("Classe r√©elle", fontsize=13)
    plt.xticks(ticks=np.arange(len(class_names)), labels=[c.split('_')[0] for c in class_names],
               rotation=90, fontsize=8)
    plt.yticks(ticks=np.arange(len(class_names)), labels=[c.split('_')[0] for c in class_names],
               rotation=0, fontsize=8)
    plt.tight_layout()
    plt.show()

    # ===== Rapport de classification =====
    print("\nüìä Rapport de classification :")
    print(classification_report(y_true, y_pred, target_names=class_names, digits=3))

    # ===== Visualisation des erreurs =====
    errors = [c for c, n in Counter(zip(y_true, y_pred)).items() if c[0] != c[1]]
    print(f"\nNombre total d'erreurs : {len(errors)} / {len(y_true)}")

    # ===== Classes les plus confondues =====
    conf_pairs = {}
    for t, p in zip(y_true, y_pred):
        if t != p:
            pair = (class_names[t], class_names[p])
            conf_pairs[pair] = conf_pairs.get(pair, 0) + 1

    top_confused = sorted(conf_pairs.items(), key=lambda x: x[1], reverse=True)[:10]
    if top_confused:
        pairs, counts = zip(*top_confused)
        labels = [f"{a} ‚Üí {b}" for a, b in pairs]
        plt.figure(figsize=(10, 5))
        plt.barh(labels, counts)
        plt.gca().invert_yaxis()
        plt.title("üîç Top 10 des confusions les plus fr√©quentes")
        plt.xlabel("Nombre d‚Äôerreurs")
        plt.tight_layout()
        plt.show()

    # ===== Pr√©cision globale =====
    acc = 100 * np.mean(np.array(y_true) == np.array(y_pred))
    print(f"\n‚úÖ Accuracy globale : {acc:.2f}%")

# === Appel de la fonction ===
visualize_model_performance(model, test_loader, class_names, device=device)
```

```python
import torch
from PIL import Image
import torchvision.transforms as T
import matplotlib.pyplot as plt
import os

def load_trained_model(weights_path, num_classes, device="cuda"):
    """
    Recharge le mod√®le HybridEfficientNetViT avec les bons poids entra√Æn√©s.
    """
    model = HybridEfficientNetViT(num_classes=num_classes, cfg=Config())
    state_dict = torch.load(weights_path, map_location=device)
    model.load_state_dict(state_dict)
    model = model.to(device)
    model.eval()
    return model

def preprocess_image(img_path):
    """
    Charge 1 image disque -> Tensor normalis√© comme le loader 'eval_tf'.
    """
    eval_tf = T.Compose([
        T.Resize((Config.img_size, Config.img_size)),
        T.ToTensor(),
        T.Normalize([0.485,0.456,0.406],[0.229,0.224,0.225]),
    ])
    img_pil = Image.open(img_path).convert("RGB")
    img_tensor = eval_tf(img_pil).unsqueeze(0)  # [1,3,H,W]
    return img_pil, img_tensor

def predict_image(model, img_tensor, class_names, device="cuda"):
    """
    Fait passer l'image dans le mod√®le et renvoie:
    - index classe pr√©dite
    - nom de la classe
    - probabilit√© softmax associ√©e
    """
    img_tensor = img_tensor.to(device)
    with torch.no_grad():
        logits = model(img_tensor)            # [1, num_classes]
        probs = torch.softmax(logits, dim=1)  # [1, num_classes]
        conf, pred_idx = torch.max(probs, dim=1)
        pred_idx = pred_idx.item()
        conf = conf.item()
    pred_class_name = class_names[pred_idx]
    return pred_idx, pred_class_name, conf, probs.squeeze().cpu().numpy()

# ============================
# 1) CONFIGURE TON IMAGE ICI
# ============================

# EXEMPLE: une image dans le dossier test, classe "tomato_early_blight"
# Mets un vrai chemin vers une image feuille (JPEG/PNG)
img_path = r"test/tomato_early_blight/tomato_early_blight_003609_6dadfbc0.png"

# ============================
# 2) CHARGEMENT DU MOD√àLE
# ============================
device = "cuda" if torch.cuda.is_available() else "cpu"

best_weights_path = os.path.join(PROJECT_ROOT, "best_hybrid_plantvillage.pth")
model_infer = load_trained_model(best_weights_path, num_classes=len(class_names), device=device)

# ============================
# 3) PR√âPARATION IMAGE + PRED
# ============================
img_pil, img_tensor = preprocess_image(img_path)
pred_idx, pred_name, conf, all_probs = predict_image(model_infer, img_tensor, class_names, device=device)

# ============================
# 4) AFFICHAGE R√âSULTAT
# ============================
plt.imshow(img_pil)
plt.axis("off")
plt.title(f"Pr√©diction: {pred_name}\nConfiance: {conf*100:.2f}%")
plt.show()

print("üî• R√©sultat inference sur une image :")
print(f"- Classe pr√©dite      : {pred_name}")
print(f"- Indice de classe    : {pred_idx}")
print(f"- Confiance (%)       : {conf*100:.2f}")
print("\nTop 5 classes probables :")

# afficher top5
import numpy as np
top5_idx = np.argsort(all_probs)[::-1][:5]
for rank, idx in enumerate(top5_idx, start=1):
    print(f"{rank}. {class_names[idx]} -> {all_probs[idx]*100:.2f}%")
```

```python
# ================== Validation crois√©e (k-fold) pour v√©rifier le surapprentissage ==================
from sklearn.model_selection import StratifiedKFold
from torch.utils.data import Subset, DataLoader
import numpy as np

k_folds = 5
batch_size_cv = 32

base_dataset = train_loader.dataset
all_labels = [base_dataset[i][1] for i in range(len(base_dataset))]
all_labels = np.array(all_labels)

skf = StratifiedKFold(n_splits=k_folds, shuffle=True, random_state=42)

acc_scores = []
f1_scores = []

for fold, (train_idx, val_idx) in enumerate(skf.split(np.zeros(len(all_labels)), all_labels), 1):
    print(f"===== Fold {fold}/{k_folds} =====")

    train_subset = Subset(base_dataset, train_idx)
    val_subset   = Subset(base_dataset, val_idx)

    train_loader_cv = DataLoader(train_subset, batch_size=batch_size_cv, shuffle=True,  num_workers=2)
    val_loader_cv   = DataLoader(val_subset,   batch_size=batch_size_cv, shuffle=False, num_workers=2)

    model_cv = HybridEfficientNetViT(num_classes=num_classes, cfg=Config()).to(device)
    optimizer = torch.optim.Adam(model_cv.parameters(), lr=1e-4)
    criterion = nn.CrossEntropyLoss()

    n_epochs_cv = 5
    for epoch in range(n_epochs_cv):
        model_cv.train()
        running_loss = 0.0
        for x, y in train_loader_cv:
            x, y = x.to(device), y.to(device)
            optimizer.zero_grad()
            outputs = model_cv(x)
            loss = criterion(outputs, y)
            loss.backward()
            optimizer.step()
            running_loss += loss.item() * x.size(0)
        epoch_loss = running_loss / len(train_loader_cv.dataset)
        print(f"  √âpoque {epoch+1}/{n_epochs_cv} - Loss train: {epoch_loss:.4f}")

    model_cv.eval()
    all_true, all_pred = [], []
    with torch.no_grad():
        for x, y in val_loader_cv:
            x, y = x.to(device), y.to(device)
            outputs = model_cv(x)
            preds = outputs.argmax(1)
            all_true.extend(y.cpu().numpy())
            all_pred.extend(preds.cpu().numpy())

    acc = accuracy_score(all_true, all_pred)
    f1 = f1_score(all_true, all_pred, average="macro")

    acc_scores.append(acc)
    f1_scores.append(f1)

    print(f"  Fold {fold} - Accuracy: {acc:.4f} | F1 macro: {f1:.4f}\n")

print("===== R√©sum√© validation crois√©e =====")
print(f"Accuracy moyenne : {np.mean(acc_scores):.4f}  ¬± {np.std(acc_scores):.4f}")
print(f"F1 macro moyen   : {np.mean(f1_scores):.4f}  ¬± {np.std(f1_scores):.4f}")
```
